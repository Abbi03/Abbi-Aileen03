```python
df= pd.read_csv('antropometria-dataset-1.csv')
df.columns
```

```python
columnP = [df.loc[:,'peso'],]
patches = plt.hist(columnP, bins = 15)
```

```python
sns.lmplot(x='sistol',
           y='diastol',
           data=df,
           fit_reg=False,
           hue= 'sexo'
          )
plt.title('Scatter plot of sistol, diastol, sexo')
```

```python
sns.boxplot(data=df.loc[:,['cintura']])
```
